### **6-1. `isdigit()`**

- 전부 “숫자 문자”면 True 나옴
- 단, `1`, `3.14` 같은 건 False 나옴(마이너스/점 때문에)
- 전각 숫자(예: `１２３`) 같은 것도 True 될 수 있음

```python
print("123".isdigit())     # True
print("-123".isdigit())    # False
print("3.14".isdigit())    # False
print("12a".isdigit())     # False
```

### **6-2. `isnumeric()` / `isdecimal()`**

- `isdecimal()` 가장 빡셈(10진 숫자만)
- `isdigit()` 그 다음
- `isnumeric()` 제일 넓음(분수/특수 숫자 문자까지 포함 가능)

```python
print("123".isdecimal())   # True
print("123".isdigit())     # True
print("123".isnumeric())   # True
```

#### **6-3. `isalpha()` / `isalnum()`**
- `isalpha()` 전부 글자면 True(공백/숫자 있으면 False)
- `isalnum()` 글자+숫자만 있으면 True(특수문자 있으면 False)

```python
print("abc".isalpha())   # True
print("abc1".isalpha())  # False

print("abc1".isalnum())  # True
print("abc_1".isalnum()) # False
```

#### **6-4. `isspace()`**

- 공백(스페이스/탭/개행)만 있으면 True

```python
print("   ".isspace())     # True
print("\n\t".isspace())    # True
print(" a ".isspace())     # False

print('\n\t')
```

## **2-2-4. bool (불리언: True/False)**

### **1) bool 기본**

- `bool`은 **True / False** 두 값만 갖는 타입이고, 비교/조건문의 결과로 가장 많이 등장.

```python
print(10 > 3)   # True
print(10 == 3)  # False
```

### **2) 논리 연산 (and, or, not)**

```python
a, b = True, False

print(a and b)  # False (둘 다 True여야 True)
print(a or b)   # True  (하나라도 True면 True)
print(not a)    # False (뒤집기)
```

### **3) Truthy / Falsy (자동으로 bool로 해석되는 규칙)**

파이썬은 조건문에서 다음 값들을 **False처럼** 취급:

- `0`, `0.0`
- `""` (빈 문자열)
- `[]`, `{}`, `set()` (빈 컨테이너)
- `None`

```python
print(bool(0))                   # False
print(bool(""))                  # False
print(bool([]))                  # False
print(bool(None))                # False
print(bool("0"), bool([0]))      # True True
```

### **4) `all()` 과 `any()`**

- 둘 다 “여러 조건을 한 번에” 판단할 때 쓰는 함수.

#### **4-1) `all(iterable)`**

- **모든 값이 True(Truthy)** 이면 True
- 하나라도 False(Falsy)면 False

```python
print(all([True, True, True]))           # True
print(all([True, False, True]))          # False
print(all([1, "hi", [0]]))               # True
print(all([1, "", 3]))                   # False
```

#### **4-2) `any(iterable)`**

- **하나라도 True(Truthy)** 이면 True
- 전부 False(Falsy)면 False


```python
print(any([False, False, True]))    # True
print(any([0, "", None]))           # False
nums = [0, 0, 5, 0]                 # True
print(any(n != 0 for n in nums))    # True
vals = [1, None, 3]
print(any(v is None for v in vals))  # True
```

### **5) `all/any` 사용할 때 중요한 포인트**

#### **비어있는 경우(빈 리스트)의 결과**

- `all([])` 는 **True**
- `any([])` 는 **False**


```python
print(all([]))  # True
print(any([]))  # False
```

### **6) not 포함**
-`not`은 **True/False를 뒤집는 연산자**.




```python
print(not True)
print(not False)
```

#### **1) 비어있다 / 비어있지 않다**
- 컨테이너(list/tuple/dict/set/str)는 **비어있으면 False**, 비어있지 않으면 True로 취급.

```python
items = []
print(not items)   # True

items = [1, 2]
print(not items)   # False

s = ""
print(not s)       # True

s = "hi"
print(not s)       # False
```

#### **2) None(값 없음) 체크**
- `None`은 “값이 없다”는 표시라서, 보통 이렇게 확인해.

```python
x = None
print(x is None)          # True

print(not (x is None))    # False
```

#### **3) “공백만 있는 문자열”도 ‘없다’로 보고 싶을 때**

- 빈 문자열이 아닌 `"   "` 같은 공백 문자열도 “없음” 처리하고 싶으면 `strip()`을 같이 씀.

```python
s = "   "
print(not s)
print(not s.strip())
```

#### **4) 포함/미포함을 not으로**

`in` 결과(True/False)에 `not`을 붙이면 “없다”가 됨.

```python
nums = [1, 2, 3]
print(not (5 in nums))     # True
print(not (2 in nums))     # False
```

```python
di = {'a':1,'b':2}

for i in di:
  print(i)         # a b
```

```python
d = {"a": 1}
print((1 in d.values()))   # True
```

#### **5) 비교 결과를 뒤집기**

```python
x = 10
print(not (x > 0))     # False
print(not (x == 10))   # False
print(not (x != 10))   # True
```

#### **6) 조건 여러 개를 not으로 묶기**

- 자주 쓰는 논리 변형.

* `not (A and B)` 는 `not A or not B` 와 같은 의미
* `not (A or B)` 는 `not A and not B` 와 같은 의미

```python
A = True
B = False

print(not (A and B))        # True
print((not A) or (not B))   # True
```
